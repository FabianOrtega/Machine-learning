# Ejemplos de Técnicas de Regresión

En el siguiente script encontrara una serie de ejemplos para trabajar técnicas de regresión con *Machine Learning*


```python
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
```

**Cargue de información**

En esta ocasión se utilizara los conjuntos de datos que maneja el modulo *scikit-learn*

```python
boston = datasets.load_boston()
# boston
```

Selección de información



```python
X = boston.data[:,5:8]
y = boston.target
```

Separar la información en datos de entrenamiento y de testeo

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Regresión Lineal

Como lo vimos en la presentación, la regresión lineal es una técnica que busca predecir el valor de una variable continua a tráves de relaciones lineales con otras variables continuas.

```python
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
y_pred = lm.predict(X_test)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(X_train,y_train)
```

## Regresión polinimial

Esta técnica permite ajustar los problemas de exclusión de una regresión lineal

```python
X_pol = boston.data[:, np.newaxis, 5]
```

```python
plt.figure()
plt.scatter(X_pol, y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pol, y, test_size=0.3, random_state=42)
```

```python
pol_rho = PolynomialFeatures(degree=2)
X_train = pol_rho.fit_transform(X_train)
X_test = pol_rho.fit_transform(X_test)
```

```python
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
y_pred = lm.predict(X_test)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(X_train,y_train)
```

**Gráfica**

```python
plt.figure()
plt.plot(X_test, y_test, color='green')
plt.plot(X_test, y_pred, color='purple', linewidth=3)
plt.show()
```
